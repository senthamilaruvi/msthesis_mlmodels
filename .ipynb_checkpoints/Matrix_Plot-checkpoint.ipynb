```python
import os
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
#Import Data
os.chdir('E:\Academics\Research\Data\Experimental data\Compression_test\Results\As_Built')
Master_Data = pd.read_csv('Data_As_Built_New.csv').dropna()
list(Master_Data)
```

```python
Input = ['X','Y','Theta','Phi','Max_Pore_Diameter','Median_Pore_Spacing']
Properties = ['Yield','Modulus','Ultimate','Ductility']
Output = Properties[0]
```

```python
# Extracting required data from master data
full = []
full.extend(Input)
full.extend(Properties)
Parsed_Data = Master_Data[full]
Parsed_Data['Max/Median']= Parsed_Data['Max_Pore_Diameter']/Parsed_Data['Median_Pore_Spacing']

# transformations
Parsed_Data['Max_Pore_Diameter'] = np.log(Parsed_Data['Max_Pore_Diameter'])
Parsed_Data = Parsed_Data[Parsed_Data['Modulus'] <100]
print "shape:", Parsed_Data.shape
```
